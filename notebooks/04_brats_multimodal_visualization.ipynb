```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

base_path = "../data/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/"

flair = nib.load(base_path + "BraTS20_Training_001_flair.nii").get_fdata()
t1 = nib.load(base_path + "BraTS20_Training_001_t1.nii").get_fdata()
t1ce = nib.load(base_path + "BraTS20_Training_001_t1ce.nii").get_fdata()
t2 = nib.load(base_path + "BraTS20_Training_001_t2.nii").get_fdata()
seg = nib.load(base_path + "BraTS20_Training_001_seg.nii").get_fdata()
```

```python
print(flair.shape, t1.shape, seg.shape)
```

```python
slice_index = flair.shape[2] // 2

plt.figure(figsize=(12, 6))

plt.subplot(1,5,1)
plt.imshow(flair[:, :, slice_index], cmap="gray")
plt.title("FLAIR")

plt.subplot(1,5,2)
plt.imshow(t1[:, :, slice_index], cmap="gray")
plt.title("T1")

plt.subplot(1,5,3)
plt.imshow(t1ce[:, :, slice_index], cmap="gray")
plt.title("T1CE")

plt.subplot(1,5,4)
plt.imshow(t2[:, :, slice_index], cmap="gray")
plt.title("T2")

plt.subplot(1,5,5)
plt.imshow(seg[:, :, slice_index])
plt.title("Segmentation")

plt.show()
```

```python
plt.imshow(flair[:, :, slice_index], cmap="gray")
plt.imshow(seg[:, :, slice_index], alpha=0.4)  # overlay
plt.title("Tumor Overlay")
plt.axis("off")
plt.show()
```
