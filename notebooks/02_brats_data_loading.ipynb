```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

flair_path = "../data/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/BraTS20_Training_001_flair.nii"
mask_path  = "../data/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/BraTS20_Training_001_seg.nii"

flair = nib.load(flair_path).get_fdata()
mask = nib.load(mask_path).get_fdata()

print("MRI shape:", flair.shape)
print("Mask shape:", mask.shape)
```

```python
from ipywidgets import interact

def show_slice(slice_index):
    
    plt.figure(figsize=(10,5))
    
    # MRI
    plt.subplot(1,2,1)
    plt.imshow(flair[:,:,slice_index], cmap="gray")
    plt.title(f"MRI Slice {slice_index}")
    plt.axis("off")
    
    # MRI + Tumor overlay
    plt.subplot(1,2,2)
    plt.imshow(flair[:,:,slice_index], cmap="gray")
    plt.imshow(mask[:,:,slice_index], cmap="jet", alpha=0.5)
    plt.title("MRI + Tumor Segmentation")
    plt.axis("off")
    
    plt.show()

interact(show_slice, slice_index=(0, flair.shape[2]-1))
```
